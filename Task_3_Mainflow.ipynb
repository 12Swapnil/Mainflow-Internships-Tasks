```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data
np.random.seed(0)
dates = pd.date_range('2023-01-01', periods=10)
data = {
    'A': np.random.rand(10),
    'B': np.random.rand(10) * 100,
    'C': np.random.rand(10) * 500
}

# Creating the DataFrame
df = pd.DataFrame(data, index=dates)
print(df)
```

```python
# Bar chart
date_to_plot = '2023-01-01'

# Select data for the chosen date
selected_data = df.loc[date_to_plot, ['B', 'C']]

# Plotting the bar chart
plt.figure(figsize=(10, 6))
selected_data.plot(kind='bar', alpha=0.75)

# Customize labels and title
plt.title('Bar Chart for {}'.format(date_to_plot))
plt.xlabel('Categories')
plt.ylabel('Values')

# Show plot
plt.legend()
plt.tight_layout()
plt.show()
```

```python
 #Line chart
plt.figure(figsize=(10, 6))
df['A'].plot(label='A', marker='o')
df['B'].plot(label='B', marker='s')

# Customize labels and title
plt.title('Line Chart for Columns A and B')
plt.xlabel('Date')
plt.ylabel('Values')

# Adding legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```
